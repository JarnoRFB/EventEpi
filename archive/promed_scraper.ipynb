```python
# pylint: disable=print-statement

import pprint
import sys; sys.path.append('./gritsapi/')
from gritsapi.scraper.scrape_promed import scrape_promed_url
import requests
import re
from pyquery import PyQuery as pq
from sys import stdout
# from __future__ import print_function

# define settings for the pretty printer and the printer itself
width = 80
pp = pprint.PrettyPrinter(indent=4, width=width)

# function to pretty print the parsed page with a delimiter if needed
def print_result(obj, print_delimiter):
    pp.pprint(parsed)
    if print_delimiter:
        print "=" * width + "\n"

def get_request(url,proxies):
    get_request = requests.get(url,proxies = proxies)
    html = get_request.json()['return'] 
    htmlDom = pq(html, parser='html_fragments')  
    return htmlDom

# define proxies to use (RKI internal)
proxies = {
    'http': 'http://fw-bln.rki.ivbb.bund.de:8020',
    'https': 'http://fw-bln.rki.ivbb.bund.de:8020',
}

url = ('https://www.promedmail.org/ajax/runSearch.php?pagenum={0}'
       '&kwby1=summary&search=&date1=02/01/2018&date2=10/24/2018'
       '&feed_id=1&submit=next'.format(0))


num_page_of = get_request(url,proxies)('p').text()
num_page_of = num_page_of.encode('ascii')
number_of_pages = re.search(r'\d+ of (\d+)',num_page_of).group(1)

for i in range(int(number_of_pages)):
    url = ('https://www.promedmail.org/ajax/runSearch.php?pagenum={0}'
       '&kwby1=summary&search=&date1=02/01/2018&date2=10/24/2018'
       '&feed_id=1&submit=next'.format(i))
    links = get_request(url,proxies)('a[id^="id"]')
    links = [link.attr['id'] for link in links.items()]
    link_ids = [re.search(r'id(\d+)',link).group(1) for link in links]
    #stdout.write("\r%d" % i + '/'+ str(number_of_pages) + ' pages processed' )
    #stdout.flush()
    print(str(i) + '/' + number_of_pages + ' pages')
    parsed = []
    for link_id in link_ids:
        parsed.append(scrape_promed_url("https://www.promedmail.org/post/" + str(link_id), proxies))
```
